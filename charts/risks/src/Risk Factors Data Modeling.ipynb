```python
import pandas as pd
import numpy as np
import datetime
```

```python
risk_factors = pd.read_csv("../data/number-of-deaths-by-risk-factor.csv")
risk_factors_df = pd.DataFrame(risk_factors)
```

```python
entity = []
year = []
count = []
factor = []
code = []
    
for i in range(0, len(risk_factors)):
    for j in range(3, len(risk_factors.columns)):
        
        factor.append(risk_factors.columns[j])
        entity.append(risk_factors.iloc[i, 0])
        code.append(risk_factors.iloc[i, 1])
        year.append(risk_factors.iloc[i, 2])
        count.append(risk_factors.iloc[i, j])

risk_factors_dict = {"Entity": entity, "Code": code, "Year": year, "Factor": factor, "Count": count}
factors_df = pd.DataFrame(risk_factors_dict)

factors_df.to_csv('../data/risk_factors_2020_03_22.csv')
```
