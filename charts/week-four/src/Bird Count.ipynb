```python
import pandas as pd
import numpy as np
import datetime
```

```python
bird_counts = pd.read_csv("../data/CBC_2013_Tin_MTN.csv")
bird_counts = pd.DataFrame(bird_counts)
```

```python
bird_counts.head(50)
```

```python
species = []
year = []
count = []
species_index = []

for i in range(0, len(bird_counts)):
    species_total = 0
    for j in range(1, len(bird_counts.columns)):
        if np.isnan(bird_counts.iloc[i, j]):
            bird_counts.iloc[i, j] = 0
        species_total = int(bird_counts.iloc[i, j]) + species_total
        #if (j == 25):
            #print(species_total)
        

for i in range(0, len(bird_counts)):
    for j in range(1, len(bird_counts.columns)):
        species.append(bird_counts.iloc[i, 0].strip())
        species_index.append(i)
        
        x = datetime.datetime(int(bird_counts.columns[j]), 1, 1)
        
        year.append(x.strftime("%y"))
                    
        count.append(bird_counts.iloc[i, j])
    

bird_counts_dict = {"Species": species, "Year": year, "Count": count, "Species_Index": species_index}
bird_counts_df = pd.DataFrame(bird_counts_dict)

bird_counts_df.to_csv('../data/bird_counts_edited.csv')
```
