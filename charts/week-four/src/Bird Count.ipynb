```python
import pandas as pd
import numpy as np
import datetime
```

```python
bird_counts = pd.read_csv("../data/CBC-data-summary-1989-2019.csv")
bird_counts = pd.DataFrame(bird_counts)
```

```python
bird_counts.head(50)
```

```python
cp_substring = "CP"
if (cp_substring in "CP"):
    print("IN THERE")
```

```python
species = []
year = []
count = []
species_index = []

for i in range(0, len(bird_counts)-1):
    species_total = 0
    for j in range(2, len(bird_counts.columns)-2):        
        year_value = str(bird_counts.iloc[i, j])
        
        cp_substring = "CP"
        if (cp_substring in year_value):
            year_value = 0.0
        else:
            year_value = float(bird_counts.iloc[i, j])
            if np.isnan(year_value):
                year_value = 0.0
                
        bird_counts.iloc[i, j] = year_value
        species_total = int(year_value) + species_total
    
for i in range(0, len(bird_counts)-1):
    for j in range(2, len(bird_counts.columns)-2):
        
        if (bird_counts.iloc[i, 1].strip() == "Total Individuals"):
            species.append("All Bird Species")
        elif(bird_counts.iloc[i, 1].strip() == "Yellow bellied Sapsucker"):
            species.append("Yellow-bellied sapsucker")
        elif(bird_counts.iloc[i, 1].strip() == "Thursh, Hermit"):
            species.append("Thrush, Hermit")
        elif(bird_counts.iloc[i, 1].strip() == "Woodpecker, Red Bellied"):
            species.append("Woodpecker, Red-bellied")
        else:
            species.append(bird_counts.iloc[i, 1].strip())
            
        species_index.append(i)
        
        x = datetime.datetime(int(bird_counts.columns[j]), 1, 1)
        
        year.append(x.strftime("%y"))
                    
        count.append(bird_counts.iloc[i, j])
    

bird_counts_dict = {"Species": species, "Year": year, "Count": count, "Species_Index": species_index}
bird_counts_df = pd.DataFrame(bird_counts_dict)

bird_counts_df.to_csv('../data/bird_counts_edited_2020_03_21.csv')
```
